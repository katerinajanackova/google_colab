```python
import seaborn as sns

# loading built-in dataset about tips in restaurant

data = sns.load_dataset('tips')

# dataset overview

data

#first chart - histogram

data['day'].hist()
```

```python
# dataset visualization - total_bill vs. tip
sns.set(style="darkgrid")
sns.relplot(x="total_bill", y="tip", data=data)
```

```python
# dataset visualization - total_bill vs. tip
sns.set(style="darkgrid")
plot = sns.relplot(x="total_bill", y="tip", hue="sex", data=data)
```

```python
# dataset visualization - total_bill vs. tip
sns.set(style="darkgrid")
plot = sns.relplot(x="total_bill", y="tip", hue="smoker", data=data)
```

```python
# save last figure

plot.figure.savefig("tips.png")
```
