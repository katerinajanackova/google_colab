```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = sns.load_dataset('titanic')
```

```python
titanic_data.tail(10)
```

```python
# nastavení vzhledu grafů na darkgrid
sns.set_style("darkgrid")
```

```python
# Histogram – počet cestujících podle třídy jízdenky (sloupec "class")

plt.figure(figsize=(8,5))
sns.histplot(data=titanic_data, x="class", stat="count", color="skyblue")
plt.title("Počet cestujících podle třídy jízdenky")
plt.xlabel("Třída jízdenky")
plt.ylabel("Počet cestujících")
plt.show()
```

```python
# Bodový graf – vztah mezi věkem a cenou jízdného, rozdělený podle pohlaví
plt.figure(figsize=(8,6))
plot = sns.scatterplot(data=titanic_data, x="age", y="fare", hue="sex")
plt.title("Vztah mezi věkem a cenou jízdného (podle pohlaví)")
plt.xlabel("Věk")
plt.ylabel("Cena jízdného")
plt.show()
```

```python
# Uložení grafu do souboru titanic.png do hlavní složky Google Colab
plot.figure.savefig("titanic.png")

# Stažení souboru do počítače
from google.colab import files
files.download("titanic.png")
```
